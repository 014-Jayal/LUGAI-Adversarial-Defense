```python
# Cell 3: Imports
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

print("Imported libraries.")
```

```python
# Cell 4: Define Paths to Saved Figures
# Paths are relative to the 'notebooks' folder
FIGURES_DIR = "../results/figures"
PURIFICATION_PLOT_PATH = os.path.join(FIGURES_DIR, "purification_visualization.png")
DETECTION_PLOT_PATH = os.path.join(FIGURES_DIR, "detection_feature_distributions.png")
```

```python
# Cell 5: Show Final Purification (Healing)
print("Displaying final purification (self-healing) result:")

try:
    img = mpimg.imread(PURIFICATION_PLOT_PATH)
    plt.figure(figsize=(15, 10))
    plt.imshow(img)
    plt.axis("off")
    plt.title("Visual Proof of Purification (Clean vs. Attacked vs. Healed)", fontsize=18)
    plt.show()
except FileNotFoundError:
    print(f"---!!! ERROR: Plot not found at {PURIFICATION_PLOT_PATH} !!!---")
    print("Please run Notebook 5 or 'python src/visualize_purification.py' first.")
```

```python
# Cell 6: Show Final Detection Features
print("\nDisplaying detection feature distributions:")

try:
    img = mpimg.imread(DETECTION_PLOT_PATH)
    plt.figure(figsize=(18, 7))
    plt.imshow(img)
    plt.axis("off")
    plt.title("Feature Distributions for Stage 1 Detection", fontsize=18)
    plt.show()
except FileNotFoundError:
    print(f"---!!! ERROR: Plot not found at {DETECTION_PLOT_PATH} !!!---")
    print("Please run Notebook 4 or 'python src/evaluate_detection.py' first.")
```
